```python
# Add Pandas dependency
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load files
avg_rain_state_to_load = "avg_rain_state.csv"

rain_df = pd.read_csv(avg_rain_state_to_load)
rain_df
```

```python
rain_df.sort_values(["Rank"], ascending=False).tail()
```

```python
x_axis = np.arange(len(rain_df))

tick_locations = [value for value in x_axis]

plt.figure(figsize=(20,3))

plt.bar(x_axis, rain_df["Inches"])

plt.xticks(tick_locations, rain_df["State"], rotation="horizontal")

plt.title("Rain in Inches by State")
plt.xlabel("State")
plt.ylabel("Rain (inches)")
plt.show()
```
